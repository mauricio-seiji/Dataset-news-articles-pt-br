```python
import openai

import os, glob

from pathlib import Path
```

```python
openai.api_key = ""
```

```python
def get_summarize(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,     
    )
    return response.choices[0].message["content"]


def read_text(filename):
    with open(filename) as f:
        return f.read()

    
def get_path(filename):
    path = Path(filename)
    directory = path.parent.name
    filename_to_save = os.path.join(directory, path.name)
    return directory, filename_to_save

def file_exists(filename):
    _, filename_to_save = get_path(filename)
    
    return os.path.isfile(filename_to_save)
    
def save_file(filename, text):
    directory, filename_to_save = get_path(filename)
    os.makedirs(directory, exist_ok=True)
    
    with open(filename_to_save, "w+") as f:
        f.write(text)
```

```python
files = glob.glob("../**/*.txt")
```

```python
summaries = []

for filename in files:    
    if file_exists(filename):
        print(f"Skipping file {filename}")
        continue
    
    text = read_text(filename)  
        
    if len(text) >= 4097:
        print(f"Skipping max token {filename}")
        continue
    
        
    print(f"Resuming file {filename}")
    prompt = f"Resuma esse texto: {text}"
    response = get_summarize(prompt)
    
    summaries.append( ( filename, response ) )
```

```python
for filename, text in summaries:
    save_file(filename, text)
```
